test데이터는 결측치가 존재하지 않으므로 작업할 것이 없음

```python
from tqdm import tqdm
import re
train_text = list(train['data'])

clear_text_list = []

for i in tqdm(range(len(train_text))):
    plain_text = str(train_text[i])

    clear_text = plain_text.replace("\\","").replace("n"," ")

    clear_text = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]', '', clear_text)
    
    clear_text_list.append(clear_text)

train['clear_text'] = clear_text_list
train
```

```python
test_text = list(test['data'])

clear_text_list = []

for i in tqdm(range(len(test_text))):
    plain_text = str(test_text[i])

    clear_text = plain_text.replace("\\","").replace("n"," ")

    clear_text = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]', '', clear_text)
    
    clear_text_list.append(clear_text)

test['clear_text'] = clear_text_list
test
```

# 3. 모델

**-1. 분석 모델 구축**

**-2. 토큰화**

```python
from konlpy.tag import Okt
from tqdm import tqdm
```

```python
okt = Okt()
```

```python
X_train = []

train_clear_text = list(train['clear_text'])

for i in tqdm(range(len(train_clear_text))):
    token_data = okt.morphs(train_clear_text[i]) #토큰화
    token_data = [word for word in token_data if len(word) > 1]
    X_train.append(token_data)
X_train[:1]
```

```python
X_test = []

test_clear_text = list(test['clear_text'])

for i in tqdm(range(len(test_clear_text))):
    token_data = okt.morphs(test_clear_text[i])
    token_data = [word for word in token_data if len(word) > 1]
    X_test.append(token_data)
X_test[:1]
```

```python
from keras.preprocessing.text import Tokenizer

max_words = 35000
tokenizer = Tokenizer(num_words = max_words)
tokenizer.fit_on_texts(X_train) 
X_train_vec = tokenizer.texts_to_sequences(X_train) 
X_test_vec = tokenizer.texts_to_sequences(X_test)
```

**-3. 모델 훈련**

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

max_len = 300
X_train_pad = pad_sequences(X_train_vec, maxlen = max_len)
X_test_pad = pad_sequences(X_test_vec, maxlen = max_len)
```

```python
from tensorflow.keras.utils import to_categorical
y_train = to_categorical(train['category'])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Embedding
```

```python
model = Sequential()
model.add(Embedding(max_words, 120))
model.add(LSTM(120))
model.add(Dense(3, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['acc'])

history = model.fit(X_train_pad, y_train, batch_size=128, epochs=15)
```

**-4. 예측 및 결과 저장**

```python
y_pred = model.predict(X_test_pad)
```

```python
import numpy as np
submission['category'] = np.argmax(y_pred, axis = -1)
submission.to_csv('submission_20003327_고민주.csv', index=False)
```
